```python
# Load diabetes dataset

from sklearn.datasets import load_diabetes

diabetes = load_diabetes(as_frame=True)
diabetes = diabetes.frame
diabetes.head()
```

```python
#1.a
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

# Load dataset diabetes load_diabetes
diabetes = load_diabetes()
X = diabetes.data
y = diabetes.target

# Split data menjadi data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Inisialisasi 3 model base
model1 = LinearRegression()
model2 = DecisionTreeRegressor(max_depth=3, random_state=0)
model3 = RandomForestRegressor(n_estimators=100, random_state=0)

# Training model base pada data latih
model1.fit(X_train, y_train)
model2.fit(X_train, y_train)
model3.fit(X_train, y_train)

# Menggunakan metode voting untuk menggabungkan prediksi dari ketiga model
y_pred1 = model1.predict(X_test)
y_pred2 = model2.predict(X_test)
y_pred3 = model3.predict(X_test)
y_pred_ensemble = np.mean([y_pred1, y_pred2, y_pred3], axis=0)

# Evaluasi performa model ensemble learning menggunakan RMSE, MAE, dan MAPE pada data uji
rmse = np.sqrt(mean_squared_error(y_test, y_pred_ensemble))
mae = mean_absolute_error(y_test, y_pred_ensemble)
mape = np.mean(np.abs((y_test - y_pred_ensemble) / y_test)) * 100

print("RMSE: ", rmse)
print("MAE: ", mae)
print("MAPE: ", mape)
```

```python
#1.b
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import confusion_matrix
import numpy as np

# Load dataset iris
iris = load_iris()
X = iris.data
y = iris.target

# Split data menjadi data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Inisialisasi 3 model base
model1 = LogisticRegression(max_iter=1000, random_state=0)
model2 = RandomForestClassifier(n_estimators=100, random_state=0)
model3 = GradientBoostingClassifier(n_estimators=100, random_state=0)

# Training model base pada data latih
model1.fit(X_train, y_train)
model2.fit(X_train, y_train)
model3.fit(X_train, y_train)

# Membuat dataset baru dari output ketiga model base pada data uji
pred1 = model1.predict(X_test)
pred2 = model2.predict(X_test)
pred3 = model3.predict(X_test)
pred_new = np.column_stack((pred1, pred2, pred3))

# Menggunakan model meta classifier untuk menggabungkan output dari ketiga model base
meta_model = LogisticRegression(max_iter=1000, random_state=0)
meta_model.fit(pred_new, y_test)

# Memprediksi label kelas pada data uji menggunakan ensemble learning dengan teknik stacking
y_pred = meta_model.predict(pred_new)

# Evaluasi performa model ensemble learning menggunakan confusion matrix pada data uji
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)
```

```python
#2
from sklearn.datasets import load_iris
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier

# Load dataset iris
iris = load_iris()
X = iris.data
y = iris.target

# Inisialisasi model base
model = RandomForestClassifier(n_estimators=100, random_state=0)

# Lakukan 5-fold cross validation pada model base
scores = cross_val_score(model, X, y, cv=5)

# Tampilkan nilai akurasi untuk setiap fold dan rata-rata akurasi
print("Accuracy Scores:", scores)
print("Mean Accuracy:", np.mean(scores))
```
